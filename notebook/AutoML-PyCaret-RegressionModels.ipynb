```python
# ライブラリーのインポート
import pandas as pd
from pycaret.regression import *

# データセットの読込み
from pycaret.datasets import get_data 
boston_data = get_data('boston')

# PyCaretを起動
exp1 = setup(boston_data, target = 'medv', ignore_features = None)
```

```python
# モデルの構築
compare_models()
```

```python
# lightgbmのモデルを確認
lightgbm = create_model('lightgbm')
```

```python
# lightgbmのモデルを最適化
lightgbm_tuned = tune_model(lightgbm, optimize = 'MAE')
```

```python
# 解析結果の確認
evaluate_model(lightgbm_tuned)
```

```python
# アンサンブル学習
lightgbm = create_model('lightgbm')
ridge = create_model('ridge')

ensemble = blend_models([lightgbm, ridge])
```
